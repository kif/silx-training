```python
N = 32
n2 = numpy.arange(N*N)
```

```python
#Figure 1: square grid
fig = figure(1, figsize=(8,8))
scatter(n2%N, n2//N, marker="o", color="orange")
title(f"N={N}")
xlabel("x")
ylabel("y")
annotate("Center: (N/2, N/2)", (N/2, N/2), xytext=(0.6*N, 0.6*N), xycoords='data', arrowprops={"width":1}, fontsize="large")
savefig("square_crystal_coordinates.png")
```

```python
#Figure 2: Circle grid
fig = figure(2, figsize=(8,8))
N_2 = N/2
x = n2%N
y = n2//N
mask = (x-N_2)**2 + (y-N_2)**2<=N_2*N_2
scatter(x, y, marker=".", color="yellow")
scatter(x[mask], y[mask], marker="o", color="orange")
title(f"Circular crystal N={N}")
xlabel("x")
ylabel("y")
annotate("Center: (N/2, N/2)", (N/2, N/2), xytext=(0.6*N, 0.6*N), xycoords='data', arrowprops={"width":1}, fontsize="large")
savefig("circular_crystal_coordinates.png")
```

```python
#Figure 3: strained crystal

# Displacement of atoms as function of the radius
def delta(radius, crystal_radius, strain_width):
    """Displacement of atoms as function of the radius"""
    return 1 + np.tanh((radius - crystal_radius) / strain_width)

fig = figure(3, figsize=(8,8))
N_2 = N/2
x = np.arange(N)
y = np.arange(N)
X = x - N_2
Y = y - N_2
U, V = np.meshgrid(X, Y)
R = numpy.sqrt(U**2+V**2)
disp = 1e-3*delta(R, N_2, 5)
mask = (U**2 + V**2)>N_2*N_2
disp[mask] = 0
title(f"Strained crystal N={N}")
xlabel("x")
ylabel("y")
annotate("Center: (N/2, N/2)", (N/2, N/2), xytext=(0.6*N, 0.6*N), xycoords='data', arrowprops={"width":1}, fontsize="large")
q = quiver(x, y, (U*disp), (V*disp), mask)
savefig("circular_crystal_strain_map2.png")
```
